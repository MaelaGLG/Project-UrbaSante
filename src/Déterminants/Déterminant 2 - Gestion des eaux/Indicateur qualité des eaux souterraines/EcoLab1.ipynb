```python
import json
import csv

# Function to extract relevant information from the JSON files
def extract_information(json_data):
    extracted_data = []

    # Loop through each site in the JSON data
    for site in json_data['sites']:
        # Extract relevant fields
        code_activite_icpe = site['description'].get('code_activite_icpe', '')
        origine_pollution = site['description']['origine_pollution'][0] if site['description'].get('origine_pollution') else ''
        type_pollution = ", ".join(site['description'].get('type_pollution', []))
        produits = ", ".join([site['description'].get(f'SP2_PROD{i}', '') for i in range(1, 26)]) + ", " + site['description'].get('SP2_PROD_A', '')
        etat_site = site.get('surveillance', {}).get('etat', [''])[0]
        code_insee = site['identification'].get('code_insee', '')  # Extracting code_insee

        # Append extracted data to the list
        extracted_data.append({
            'code_activite_icpe': code_activite_icpe,
            'origine_pollution': origine_pollution,
            'type_pollution': type_pollution,
            'produits': produits,
            'etat_site': etat_site,
            'code_insee': code_insee  # Adding code_insee to the extracted data
        })

    return extracted_data

# Read the JSON files
json_files = ['/content/drive/MyDrive/BASOL.json']  # Update with your file paths

# Extract information from each JSON file
all_extracted_data = []
for json_file in json_files:
    with open(json_file, 'r', encoding='utf-8') as file:
        json_data = json.load(file)
        extracted_data = extract_information(json_data)
        all_extracted_data.extend(extracted_data)

# Print the extracted data (for verification)
for data in all_extracted_data:
    print(data)
```

```python
# Specify the CSV file path
csv_file = 'extracted_data.csv'

# Write the extracted data to the CSV file
with open(csv_file, 'w', newline='', encoding='latin-1') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=all_extracted_data[0].keys())
    writer.writeheader()
    writer.writerows(extracted_data)
```

```python
from collections import Counter

# Extract unique values of etat_site and count occurrences
etat_site_counter = Counter(data['etat_site'] for data in all_extracted_data)

# Print the unique values and their counts
for state, count in etat_site_counter.items():
    print(f"{state}: {count}")
```

```python
# Define weights based on the state of the site
weight_mapping = {
    'Absence de surveillance justifiée': 1,
    'Surveillance différée en raison de procédure en cours': 0.5,
}

# Function to calculate the weighted score
def calculate_weighted_score(data, weight_mapping):
    weighted_score = 0
    for entry in data:
        # Get the weight based on the state of the site
        weight = weight_mapping.get(entry['etat_site'], 0)
        # Get the count of the state
        count = etat_site_counter.get(entry['etat_site'], 0)
        # Calculate the weighted score by multiplying weight with count
        weighted_score += weight * count
    return weighted_score

# Calculate the weighted score for waste storage facilities and tanks
weighted_score = calculate_weighted_score(all_extracted_data, weight_mapping)
print("Weighted Score:", weighted_score)
```

```python
import geopandas as gpd
from shapely.geometry import Point
import json

# Function to extract 'x', 'y', and 'cd_insee' from the first JSON file
def extract_xy(json_file):
    with open(json_file, 'r') as file:
        json_data = json.load(file)

    xy_data = []
    for feature in json_data['features']:
        properties = feature['properties']
        geometry = feature['geometry']
        x = properties.get('x', None)
        y = properties.get('y', None)
        cd_insee = properties.get('cd_insee', None)  # Extract 'cd_insee' field

        if x is not None and y is not None and cd_insee is not None:
            xy_data.append({
                'x': x,
                'y': y,
                'code_insee': cd_insee  # Add 'cd_insee' to extracted data
            })

    return xy_data


# Extract 'x' and 'y' from the first JSON file
xy_data = extract_xy('/content/drive/MyDrive/icpe.geojson')

# Combine 'x' and 'y' data into a single GeoDataFrame
xy_geometry = [Point(float(data['x']), float(data['y'])) for data in xy_data]
xy_gdf = gpd.GeoDataFrame(xy_data, geometry=xy_geometry, crs='EPSG:4326')

csv_file = pd.read_csv('/content/extracted_data.csv', encoding='latin-1')

# Merge the two DataFrames based on the 'code_insee' column
sites_gdf = pd.merge(xy_gdf, csv_file, on='code_insee')

# Save the merged DataFrame to a CSV file
sites_gdf.to_csv('/content/drive/MyDrive/merged_data.csv', index=False)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Load the GeoJSON file containing population center data
population_geojson_file = '/content/drive/MyDrive/icpe.geojson'
population_centers_gdf = gpd.read_file(population_geojson_file)

# Convert the population centers to a GeoDataFrame
population_points = [Point(xy) for xy in zip(population_centers_gdf.geometry.x, population_centers_gdf.geometry.y)]
population_gdf = gpd.GeoDataFrame(population_centers_gdf, geometry=population_points)

# Re-project the population GeoDataFrame to a projected CRS
projected_crs = "EPSG:2154"  # Replace XXXX with the appropriate EPSG code for a suitable projected CRS
population_gdf = population_gdf.to_crs(projected_crs)

# Re-project the sites GeoDataFrame to the same projected CRS
sites_gdf = sites_gdf.to_crs(projected_crs)

# Calculate distances from each site to population centers
sites_gdf['distance_to_population'] = sites_gdf.geometry.apply(lambda x: population_gdf.distance(x).min())

# Define distance categories and corresponding weights
distance_categories = [(0, 250), (250, 500), (500, float('inf'))]  # Distance categories in meters
distance_weights = [1, 0.5, 0.25]  # Corresponding weights for each category

# Function to assign weights based on distance categories
def assign_weight(distance):
    for i, (min_dist, max_dist) in enumerate(distance_categories):
        if min_dist <= distance <= max_dist:
            return distance_weights[i]
    return 0

# Define weights based on the state of the site
weight_mapping = {
    'Absence de surveillance justifiée': 1,
    'Surveillance différée en raison de procédure en cours': 0.5,
}

# Function to calculate the weighted score for each site based on its state
def calculate_weighted_score(site_state, weight_mapping):
    return weight_mapping.get(site_state, 0)

# Calculate the weighted score for each site
sites_gdf['weighted_score'] = sites_gdf['etat_site'].apply(lambda x: calculate_weighted_score(x, weight_mapping))

# Apply the function to calculate weights based on distances
sites_gdf['population_weight'] = sites_gdf['distance_to_population'].apply(assign_weight)

# Multiply the weighted score of each site by its adjusted weight
sites_gdf['weighted_adjusted_score'] = sites_gdf['weighted_score'] * sites_gdf['population_weight']

# Aggregate scores based on the weighted adjusted scores
final_aggregated_score = sites_gdf['weighted_adjusted_score'].sum()

print("Final Aggregated Score:", final_aggregated_score)
```
