```python
# Combining all regional indicator dataframes into 1 single df

# List of DataFrames to concatenate
dfs_to_concat = [df_AIR_BREIZH, df_AIR_PAYS_DE_LA_LOIRE, df_AIRPARIF, df_ATMO_ARA, df_ATMO_BFC, df_ATMO_GRAND_EST,
                df_ATMO_HAUTS_DE_FRANCE, df_ATMO_NORMANDIE, df_ATMO_NOUVELLE_AQUITAINE, df_ATMO_OCCITANIE, df_ATMO_SUD,
                df_LIGAIR, df_QUALITAIR_CORSE]

# Concatenate the DataFrames along the rows
combined_df = pd.concat(dfs_to_concat, ignore_index=True)

print(combined_df)
```

```python
# Downloading the final indicator df to csv
combined_df.to_csv('Combined_df_SOMO35.csv', index=False, encoding = 'latin1')
```

```python
# Create a Basemap instance for France
map_france = Basemap(projection='merc', lat_0=0, lon_0=0,
                     llcrnrlat=41, urcrnrlat=51,
                     llcrnrlon=-5, urcrnrlon=10,
                     resolution='i')

# Draw coastlines, countries, and states
map_france.drawcoastlines()
map_france.drawcountries()
map_france.drawstates()

# Plot points with color gradient based on scores
for index, row in combined_df.iterrows():
    x, y = map_france(row['Longitude'], row['Latitude'])
    map_france.scatter(x, y, c=row['SOMO35 Subscore_x'], cmap='viridis', s=50, alpha=0.75, vmin = 0, vmax = 15000)

# Add a colorbar to show the score gradient
plt.colorbar(label='SOMO35 Score 2023')

# Show the plot
plt.show()
```
